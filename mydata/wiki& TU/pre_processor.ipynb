```python
def clean(text):
    # regex = r"([A-z])+|[0-9]|\||-|∆|”|“|’|‘|*|…|([!?\\\/\(\)\+#&])+"
    regex = r"\||-|∆|”|“|’|‘|\*|…|([!?:;'\\\/\(\)\-\+,#&])+"
    text= re.sub(regex," ",text)
    text= re.sub(r"([.])"," . ",text)
    return " ".join(text.split())
    
```

```python
def decodeDataset(directory="output1"):
    ##create dataframes for each
    with open("/content/summary.json","r") as f:
          summary =json.load(f)
    with open("/content/links.json","r") as f:
          links =json.load(f)
    count=0
    datas=[]

    for file in os.listdir(directory):
        count=count+1
        if count%10==0:
            print("--------------"+file+"--------"+str(count)+"---------")

        subDir=directory+"/"+file

        with open(subDir,"r") as f:
          data =json.load(f)
          data["summary"]=summary[data["name"]]

          datas.append(data)
        # df["summary"]=summary[data["name"]]
    df=pd.json_normalize(datas)

    return df
```

# Create stemmer mapping

```python
df=pd.read_csv("processed/playbackSingers.csv")
```

```python
df.columns.tolist()
```

```python
columns=df.columns.tolist()
columns.remove("url_TU")
columns.remove("url_WIKI")
```

```python
columns
```

```python
import snowballstemmer
```

```python
stemmer = snowballstemmer.stemmer('tamil');
```

```python
stemSet={}
for col in columns:
    ll=df[col].tolist()
    ll=set(ll)
    ll.discard(np.nan)
    ll=list(ll)
    ll=" ".join(ll)
    ll=ll.replace(".", " ")
    ll=ll.split()
    for el in ll:
        if not(el in stemSet):
            st=stemmer.stemWord(el)
            stemSet[el]=st
```

```python
stemoutput=""
for key,value in stemSet.items():
    stemoutput+=key+" => "+value+"\n"
```

```python
with open("processed/stem.txt","w",encoding="utf-8-sig") as f:
    f.write(stemoutput)
```
