```python
!pip install wikipedia -q

import json 
from bs4 import BeautifulSoup
import wikipedia
import requests
import time

for key ,value in links.items():
  value="https://ta.wikipedia.org"+value
  html=requests.get(value)
  b = BeautifulSoup(html.text, 'lxml')

  body=b.find("div", {"id": "bodyContent"}).find("div", {"id": "mw-content-text"})
  table=body.find("table", {"class": "infobox vcard"})

  Table={}
  if (table):
    for tr in table.findAll(name="tr"):
      try:
        th=tr.find(name="th").text
        td=tr.find(name="td")
        # print(th,td)
        if (th and td):
          li=td.find(name="li")
          if (li):
            td=",".join([i.text for i in td.findAll(name="li")])
          else:
            td=td.text
          Table[th]=td
      except:
        continue

  content=body.find("div", {"class": "mw-parser-output"}).findAll(["h2","p","ul"])
  if (content):
    prevTag="p"
    prevh2=""
    headercontent=""
    Page={}
    for i in content:
      t=i.name
      if t=="h2":
        # print(i)
        h2=i.find("span",{"class":"mw-headline"})
        if (h2):
          Page[prevh2]=headercontent
          headercontent=""
          prevh2=h2.text
          prevTag="h2"
          continue
      if (prevTag=="h2"):
        if (prevTag=="h2" and prevh2!=""):
          if ("ul"):
            headercontent+="Nilaanul"+i.text+"Nilaanul"
          if(t=="p"):
            headercontent+="Nilaanp"+i.text+"Nilaanp"
  dictionary={"table":Table,"page":Page,"url":value,"name":key}
  with open("/content/output1/"+key.replace("/","")+".json", "w") as outfile:
    json.dump(dictionary, outfile,ensure_ascii=False)
  print(key)
  time.sleep(3)
```

```python
dfFix=dfFix.groupby('page_url').agg({'table':"sum","top_songs":"sum","songs":"sum","summary":"first","image_url":"first"}).reset_index()
```
